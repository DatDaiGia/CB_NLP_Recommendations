```python
df = pd.read_csv('https://query.data.world/s/uikepcpffyo2nhig52xxeevdialfl7')
df
```

```python
features = ['Title', 'Year', 'Genre', 'Director', 'Actors', 'Plot']
```

```python
df = df[features]
df['Year'] = df['Year'].astype('str')
df
```

## Clean data

```python
def combined_feature(row):
    try:
        return row['Title'] + ' ' + row['Year'] + ' ' + row['Genre'] + ' ' + row['Director'] + ' ' + row['Actors'] + ' ' + row['Plot']
    except:
        print('Error' + row)
```

```python
df['new_features'] = df.apply(combined_feature, axis=1)
```

```python
df
```

```python
df['keywords'] = ''
r = Rake()
```

```python
for ind, row in df.iterrows():
    new_features = row['new_features']
    r.extract_keywords_from_text(new_features)
    keywords_dict_scores = r.get_word_degrees()
    row['keywords'] = list(keywords_dict_scores.keys())
    
df
```

```python
df['keywords'] = df['keywords'].apply(lambda x: ', '.join(x))
df
```

## Cosine similarity calculation

```python
tf_idf = TfidfVectorizer()
```

```python
count_matrix = tf_idf.fit_transform(df['keywords'])
```

```python
matrix_cosine_similarity = cosine_similarity(count_matrix, count_matrix)
```

```python
matrix_cosine_similarity
```

## Recommendations

```python
indices = pd.Series(df.index)
indices
```

```python
def recommendations(title, cosine_sim=matrix_cosine_similarity):
#     index = df[df['Title'] == title]['Title'].values(0)
#     sorted_list = sorted()
    recommend_movies = []
    idx = indices[indices==title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    print(score_series)
    top_10_indexes = list(score_series.iloc[1:11].index)
    for i in top_10_indexes:
        recommend_movies.append(list(df.index)[i])
        
    return recommend_movies
```

```python
recommendations(1)
```
